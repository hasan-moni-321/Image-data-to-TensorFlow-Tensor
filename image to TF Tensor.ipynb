```python
import numpy as np 

import os 
from tqdm import tqdm

import tensorflow as tf
from tensorflow.keras.utils import to_categorical
```

```python
def test_data_processing(image_path, image_list, data_class):

  images = [] 
  labels = [] 

  for c in data_class:
    for img_name in tqdm(os.listdir(image_path + "/" + c)):
      full_path = image_path + c + "/"+ img_name
      img = tf.io.read_file(full_path) 
      tensor = tf.io.decode_image(img, channels=3, dtype=tf.dtypes.float32)
      tensor = tf.image.resize(tensor, [128, 128])
      images.append(tensor) 
      if(c == "adidas"):
        labels.append(0)
      else: 
        labels.append(1) 

  images_array = np.array(images)
  y_categorical = to_categorical(labels) 
  y_test = np.argmax(np.round(y_categorical), axis=1)

  return images_array, y_test

image_path = "/content/drive/MyDrive/Written_Task/technical_test_venturas/dataset/test/"
image_list = os.listdir(image_path) 
class_ = ['adidas', 'nike'] 
 
x_test, y_test = test_data_processing(image_path, image_list, class_) 

print("\n\n\n")
print("Shape of x_test is: {} Shape of y_test is: {}".format(x_test.shape, y_test.shape))  
```
